#### Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns
```

```python
plt.title('Duranción segundos')
plt.hist(df['duration_seconds'],edgecolor = 'black',linewidth=1)
plt.show()
```

```python
plt.boxplot(df['duration_seconds'],vert=False) #vert=False es para que salga en horizontal
plt.show()
```

Para obtener los valores de los outliers tenemos que calcular los valores del primer y tercer cuartil

```python
Q1 = df['duration_seconds'].quantile(0.25)
print("Primer cuartil",Q1)

Q3 = df['duration_seconds'].quantile(0.75)
print("Tercer cuartil",Q3)

rango_intercuartil = Q3-Q1
print("Rango Intercuartil", rango_intercuartil)

Mediana = df['duration_seconds'].median()
print('Mediana', Mediana)

Valor_Minimo = df['duration_seconds'].min()
print("Valor Mínimo", Valor_Minimo)

Valor_Maximo = df['duration_seconds'].max()
print("Valor Máximo",Valor_Maximo)
```

Calcular los valores de los bigotes inferior y superior

```python
binferior_calculado = (Q1-1.5*rango_intercuartil)
binferior_calculado
```

```python
bsuperior_calculado = (Q3+1.5*rango_intercuartil)
bsuperior_calculado
```

Ubicar las filas en las que se encuentran los outliers

```python
ubicacion_outliers = (df['duration_seconds'] < binferior_calculado) | (df["duration_seconds"]> bsuperior_calculado)
ubicacion_outliers

# cuando es outlier es True
```

Imprimir los valores de los outliers

```python
outliers = df[ubicacion_outliers]
print("\n Lista de Outliers \n", outliers)
```

Ordenar los outliers (ascendente)

```python
outliers_ordenados = outliers.sort_values("duration_seconds")
outliers_ordenados
```

Los que no son outiliers

```python
ubicacion_sin_outliers = (df['duration_seconds'] >= binferior_calculado) & (df["duration_seconds"] <= bsuperior_calculado)
ubicacion_sin_outliers = df[ubicacion_sin_outliers]
ubicacion_sin_outliers
```

```python
plt.title('Sin outliers')
plt.boxplot(ubicacion_sin_outliers['duration_seconds'],vert=False)

plt.show()
```
